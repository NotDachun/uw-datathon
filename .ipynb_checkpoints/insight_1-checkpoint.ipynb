```python
%matplotlib inline
import pandas as pd
import matplotlib
df = pd.read_csv('kickstarter201801.csv')
df
```

```python
categories = df['category'].unique()
print(categories)
print(len(categories))
```

```python
categoryList = []
backers = []
pledged = []
for category in categories:
    newDf = df.loc[df['category'] == category]
    totalBackers = 0
    totalMoney = 0
    count = 0
    for index, row in newDf.iterrows():
        totalBackers += row['backers']
        totalMoney += row['usd_pledged_real']
        count += 1  
    categoryList.append(category)
    backers.append(totalBackers / count)
    pledged.append(totalMoney / count)
    total = 0
d = {'Category':categoryList, 'Average Backers':backers, 'Average Pledge Amount':pledged}
result = pd.DataFrame(d)
result
```

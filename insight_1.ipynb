```python
%matplotlib inline
import pandas as pd
import matplotlib
df = pd.read_csv('kickstarter201801.csv')
df
```

```python
categories = df['main_category'].unique()
print(categories)
print(len(categories))
```

```python
categoryList = []
backers = []
pledged = []
for category in categories:
    newDf = df.loc[df['main_category'] == category]
    totalBackers = 0
    totalMoney = 0
    count = 0
    for index, row in newDf.iterrows():
        totalBackers += row['backers']
        totalMoney += row['usd_pledged_real']
        count += 1  
    categoryList.append(category)
    backers.append(totalBackers / count)
    pledged.append(totalMoney / count)
    total = 0
d1 = {'Category':categoryList, 'Average Backers':backers}
d2 = {'Category':categoryList, 'Average Pledge Amount':pledged}
result1 = pd.DataFrame(d1)
result2 = pd.DataFrame(d2)
```

```python
result1.plot.bar(x='Category', y='Average Backers', title='Kickstarter - Average Backers/Project by Category')
result2.plot.bar(x='Category', y='Average Pledge Amount', title='Kickstarter - Average Pledge Amount/Project by Category')
```
